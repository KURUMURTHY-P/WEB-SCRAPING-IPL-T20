```python
# importing the required libraries
import pandas as pd

import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
# Required url
url ="https://www.iplt20.com/stats/2021"
```

```python
page =requests.get(url)
```

```python
page.status_code
```

```python
html_code =page.text
```

```python
cric = BeautifulSoup(html_code)
```

```python
cric.prettify
```

```python
player=[]
mat=[]
inns=[]
overs=[]
runs =[]
wickets =[]
bbi =[]
avg = []
econ =[]
sr =[]
wik4=[]
wik5 =[]
season =[]
for i in range(2013,2021):
    url ="https://www.iplt20.com/stats/{}/mostWkts?stats_type=bowling".format(i)
    page =requests.get(url)
    html_code =page.text
    cric =BeautifulSoup(html_code)
    #print(url)
    

    table = cric.find_all('table',class_='np-mostruns_table')
    for row in table[1].find_all('tr'):
    
        col = row.find_all('td')
        if len(col)==13:
            player.append(col[1].text.replace("\n",""))
            mat.append(col[2].text)
            inns.append(col[3].text)
            overs.append(col[4].text)
            runs.append(col[5].text)
            wickets.append(col[6].text)
            bbi.append(col[7].text)
            avg.append(col[8].text)
            econ.append(col[9].text)
            sr.append(col[10].text)
            wik4.append(col[11].text)
            wik5.append(col[12].text)
            season.append(i)
        
```

```python
import pandas as pd
df = pd.DataFrame({"player":player,
                   "matches":mat,
                   "inns":inns,                              
                   "overs":overs,
                   "runs":runs,
                   "wickets":wickets,
                   "avg":avg,
                   "bbi":bbi,
                    "sr":sr,
                   "econ":econ,
                   "wic_4":wik4,
                   "wic_5":wik5,
                   "season":season
                     })
df.head()
```

```python
df.shape
```

```python
df.info()
```

# EDA

```python
df.matches =df.matches.astype(int)
df.inns =df.inns.astype(int)
df.overs =df.overs.astype(float)
df.runs =df.runs.astype(int)
df.wickets =df.wickets.astype(int)
df.avg =df.avg.astype(float)
df.bbi =df.bbi.astype(str)
df.sr =df.sr.astype(float)
df.econ =df.econ.astype(float)
df.wic_4 =df.wic_4.astype(int)
df.wic_5 =df.wic_5.astype(int)
df.season =df.season.astype(str)
```

```python
df.info()
```

```python
best_wick =[]
runs_given =[]
lst =df.bbi.apply(lambda x :x.split("/"))
for i in lst:
    best_wick.append(i[0])
    runs_given.append(i[1])
```
