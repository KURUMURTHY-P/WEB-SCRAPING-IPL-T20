```python
# importing the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Required url
url ="https://www.iplt20.com/stats/2021"
```

```python
# Getting the url page
page =requests.get(url)
page.status_code
```

```python
# all the text in the url page
html_code =page.text
```

```python
cric = BeautifulSoup(html_code)
cric
```

```python
# getting the all html code in the proper way (in sequence)
cric.prettify
```

```python
player=[]
mat=[]
inns=[]
not_out=[]
runs =[]
high_score =[]
avg = []
balls_faced =[]
strike_rate =[]
century =[]
half_cen =[]
four =[]
six =[]
season =[]
for i in range(2013,2021):
    url ="https://www.iplt20.com/stats/{}".format(i)
    page =requests.get(url)
    html_code =page.text
    cric =BeautifulSoup(html_code)
    

    table = cric.find_all('table',class_='np-mostruns_table')
    for row in table[0].find_all('tr'):
    
        col = row.find_all('td')
        if len(col)==14:
            player.append(col[1].text.replace("\n",""))
            mat.append(col[2].text)
            inns.append(col[3].text)
            not_out.append(col[4].text)
            runs.append(col[5].text)
            high_score.append(col[6].text)
            avg.append(col[7].text)
            balls_faced.append(col[8].text)
            strike_rate.append(col[9].text)
            century.append(col[10].text)
            half_cen.append(col[11].text)
            four.append(col[12].text)
            six.append(col[13].text)
            season.append(i)
        
```

```python
import pandas as pd
df = pd.DataFrame({"player":player,
                   "matches":mat,
                   "inns":inns,
                   "not_out":not_out,
                   "runs":runs,
                   "high_score":high_score,
                   "avg":avg,
                   "balls_faced":balls_faced,
                   "strike_rate":strike_rate,
                   "century":century,
                   "half_century":half_cen,
                   "four":four,
                   "six":six,
                   "season":season
                     })
df
```
